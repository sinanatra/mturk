```python
from geopy.geocoders import Nominatim
import time
from pprint import pprint
import csv
import urllib.request
from PIL import Image
app = Nominatim(user_agent="getlocation")
```

### Geolocate

```python
csvFile = "windows photos.csv"
```

```python
f = open("out.tsv", "w")
f.write("img\tcity\tLatitude\tLongitude\n")

with open(csvFile) as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    for row in csv_reader:
        if line_count == 0:
            print(f'Column names are {", ".join(row)}')
            line_count += 1
        else:   
            photo = row[1]
            # geolocate
            city = row[2]
            location = app.geocode(city, addressdetails=True).raw
            lon = location["lon"]
            lat = location["lat"]

            #print(location['address'])
            #zipText = location['address']['postcode']
            cityText = "Undefined"
            
            if "city" in location['address']:
                cityText = location['address']['city']
            elif "town" in location['address']:
                cityText = location['address']['town']
            elif "village" in location['address']:
                cityText = location['address']['village']
                
            countryText = location['address']['country_code']

            locText = cityText +", "+countryText.upper()
            print(locText)
    
            f.write(photo+"\t"+locText+"\t"+lat+"\t"+lon+"\n")
            line_count += 1
    print(f'Processed {line_count} lines.')

f.close()
```
